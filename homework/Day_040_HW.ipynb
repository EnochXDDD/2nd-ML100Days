## 練習時間
試著使用 sklearn datasets 的其他資料集 (boston, ...)，來訓練自己的線性迴歸模型，並加上適當的正則話來觀察訓練情形。

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

boston = datasets.load_boston()
x_train, x_test, y_train, y_test = train_test_split(boston.data, boston.target, test_size=0.2, random_state=4)
regr = linear_model.LinearRegression()
regr.fit(x_train, y_train)
print(regr.coef_)
print("Mean squared error: %.2f"% mean_squared_error(y_test, regr.predict(x_test)))
```

# Lasso

```python
MSE_list = []
alpha_list = list(map(lambda e: 10.0 ** e, range(0, -10, -1)))
for alpha in alpha_list:
    lasso = linear_model.Lasso(alpha=alpha)
    lasso.fit(x_train, y_train)
    MSE_list.append(mean_squared_error(y_test, lasso.predict(x_test)))
plt.plot(alpha_list, MSE_list)
plt.title("min: {}".format(min(MSE_list)))
plt.show()
```

# Ridge

```python
MSE_list = []
alpha_list = list(map(lambda e: 10.0 ** e, range(0, -10, -1)))
for alpha in alpha_list:
    ridge = linear_model.Ridge(alpha=alpha)
    ridge.fit(x_train, y_train)
    MSE_list.append(mean_squared_error(y_test, ridge.predict(x_test)))
plt.plot(alpha_list, MSE_list)
plt.title("min: {}".format(min(MSE_list)))
plt.show()
```
