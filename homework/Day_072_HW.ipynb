# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline

def ReLU(x):
    return max(0, x)

def dReLU(x):
    return float(x > 0)

x = np.linspace(-10, 10, 201)

plt.figure()
plt.plot(x, list(map(ReLU, x)))

plt.figure()
plt.plot(x, list(map(dReLU, x)))

plt.show()
```
